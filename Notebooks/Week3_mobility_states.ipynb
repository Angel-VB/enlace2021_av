# Mobility trends of Mexican States during 2020-2021


This Notebook uses the mobility dataframe for the Mexican region and it's States (admin1) derived from the 2020-2021 Community Mobility Reports : [GOOGLE](https://www.google.com/covid19/mobility/?hl=en).

```python
import pandas as pd
```

Mobility report for mexican mobility trends in 2020

```python
df2020 = pd.read_csv('../data/2020_MX_Region_Mobility_Report.csv')
df2020
```

## Mobility report for mexican mobility trends in 2021

The dataframe of 2020 mexican mobility is read

```python
df2021 = pd.read_csv('../data/2021_MX_Region_Mobility_Report.csv')
df2021
```

The whole 2020 country mobility trends is excluded leaving only it's states 

```python
df2020 = df2020.loc[321:,:].copy()
df2020
```

## Mobility report for mexican mobility trends in 2021

The dataframe of 2021 mexican mobility is read

```python
df2021 = df2021.loc[198:,:].copy()
df2021
```

## Computing averages of mobility trends per State

A subset cotaining only the name of the states is created

```python
states = df2020.drop_duplicates(subset=['sub_region_1'])['sub_region_1'].copy()
```

A new dataframe that will contain all the data of intereset for further analyzes is created

```python
dfAll = pd.DataFrame(columns = ['retail_and_recreation_mobility', 
                                'grocery_and_pharmacy_mobility',
                                'parks_percent_mobility',
                                'transit_stations_mobility',
                                'workplaces_mobility',
                                'residential_mobility'])
```

Using a design algorithm the averages of the different mobility trends in each state during is calculated in the period of 2020-2021.

Since the positivity of the data means that theres a lower mobility and viceversa with the negativity, the data was multiplied by minus one to have a better correlation fitting data.

```python
for region in states:
    dfAll2=pd.DataFrame()
    query = "sub_region_1 == '{}'".format(region)
    df=df2020.query(query).copy()
    df2=df2021.query(query).copy()
    days2020, var = df.shape
    days2021, var = df2.shape
    dfAll2=[-1*(df['retail_and_recreation_percent_change_from_baseline'].sum()+df2['retail_and_recreation_percent_change_from_baseline'].sum())/(days2020+days2021),
            -1*(df['grocery_and_pharmacy_percent_change_from_baseline'].sum()+df2['grocery_and_pharmacy_percent_change_from_baseline'].sum())/(days2020+days2021),
            -1*(df['parks_percent_change_from_baseline'].sum()+df2['parks_percent_change_from_baseline'].sum())/(days2020+days2021),
            -1*(df['transit_stations_percent_change_from_baseline'].sum()+df2['transit_stations_percent_change_from_baseline'].sum())/(days2020+days2021),
            -1*(df['workplaces_percent_change_from_baseline'].sum()+df2['workplaces_percent_change_from_baseline'].sum())/(days2020+days2021),
            -1*(df['residential_percent_change_from_baseline'].sum()+df2['residential_percent_change_from_baseline'].sum())/(days2020+days2021)]
    a_series = pd.Series(dfAll2, index = dfAll.columns)
    dfAll = dfAll.append(a_series, ignore_index=True).copy()
dfAll.head()
```

Through cleaning and formatting the state name data a new row was included to the mobility data.

```python
dfAll2 = pd.DataFrame(columns =['state'])
dfAll2['state'] = states.str.upper().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
dfAll2 = dfAll2.set_index(dfAll.index)
dfAll['state'] = dfAll2['state']
dfAll.at[13,'state']='DISTRITO FEDERAL'
dfAll.at[25,'state']='MEXICO'
dfAll.head()
```

The week 2 institutions states code dataframe is read so that the data of the mobility can be easily manipulated.

```python
dfStates = pd.read_csv('../data/week2_states_code_institution.csv',usecols=['cve_ent','state_conacyt'])
dfStates.rename(columns={'state_conacyt':'state'},inplace=True)
dfStates.head()
```

By using the states codes data and the mobility data the final dataframe including states and the state code is included so the data can be easily differentiable by state.

```python
dfFinal = pd.merge(dfStates,dfAll,on=['state'])
dfFinal.head()
```

The dataframe is stored

```python
dfFinal.to_csv('../data/week3_mobility_states.csv',index=False)
```
